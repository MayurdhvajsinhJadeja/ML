## Question 1

```python
# posts_short['tags'].value_counts()
```

```python
cnt = []
count = 0
for i in tags:
    j = i.replace('><',' ')
    ls = list(j.split())
    cnt.append(len(ls))
    
print(max(cnt))
print(min(cnt))
x = range(1,max(cnt)+1)
y = []
for i in x:
    y.append(cnt.count(i))
print(y)
```

```python
import matplotlib.pyplot as plt
plt.bar(x,y)
plt.xlabel('No. of tags in a single row')
plt.ylabel('Frequency of occurance of No. of tags')
```

## Question 2

```python
tags_col = posts_short['tags'].tolist()
```

```python
tags_col_str = ""
s = tags_col_str.join(tags_col)
s = s.replace("<","")
s = s.replace(">"," ")
ls = s.split()
```

```python
# print(ls)
```

```python
x = np.unique(ls)
print("Total Number of Unique Tags: ",len(x))
```

## Question 3

```python
import collections
from operator import itemgetter
frequencyDict = collections.Counter(ls)
# print(frequencyDict)
top25 = dict(sorted(frequencyDict.items(), key = itemgetter(1), reverse = True)[:25])
print(top25)
```

```python
names25 = list(top25.keys())
values25 = list(top25.values())
c = [{v25<60000: 'y', v25>=60000: 'g'}[True] for v25 in values25]
plt.figure(figsize=(20,10))
plt.bar(range(len(top25)), values25, tick_label= names25, color=c)
```

## Question 9

```python
covered_que = 0
total_que = 0

for i in tags:
    total_que += 1
    temp = i.split('>')
    temp.pop()
    tags_current_que = len(temp)
    covered_tags = 0
    
    for j in temp:
        if j[1:] in names500:
            covered_tags += 1
    if covered_tags == tags_current_que:
        covered_que += 1
    
print(total_que)
print('No. of Questions Covered by Top 500 tags', (covered_que/total_que)*100)
```
